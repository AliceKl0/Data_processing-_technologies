5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
recipes3 = recipes.rename(columns={'id': 'recipe_id'})

recipes_and_reviews2 = pd.merge(reviews['recipe_id'], recipes3[['recipe_id', 'name']], how='right')

review_count_column = reviews.groupby('recipe_id').size().reset_index(name='review_count')

recipes_and_reviews2 = pd.merge(recipes_and_reviews2, review_count_column, on='recipe_id', how='left')

unique_recipes_and_reviews2 = recipes_and_reviews2.drop_duplicates()
unique_recipes_and_reviews2 = unique_recipes_and_reviews2.fillna(0)
unique_recipes_and_reviews2
```

```python
unique_recipes_and_reviews2[unique_recipes_and_reviews2.recipe_id==363362] # Проверка (действительно, у рецепта с id 363362 0 отзывов) 
```
